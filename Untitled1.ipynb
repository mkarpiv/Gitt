# Робота з Python. Найпростіший калькулятор

Спочатку вам запропоновано вибрати дію (додавання, віднімання, множення або ділення), а потім ввести два числа. Після введення чисел калькулятор виконає вибрану операцію та відобразить результат.

```python

def add(x, y):
    return x + y

def subtract(x, y):
    return x - y

def multiply(x, y):
    return x * y

def divide(x, y):
    if y != 0:
        return x / y
    else:
        return "Ділення на нуль не допускається."

def calculator():
    print("Select operation:")
    print("1. Додавання")
    print("2. Віднімання")
    print("3. Множення")
    print("4. Ділення")

    choice = input("Enter choice (1/2/3/4): ")

    num1 = float(input("Введіть перше число: "))
    num2 = float(input("Введіть друге число: "))

    if choice == '1':
        print("Result: ", add(num1, num2))
    elif choice == '2':
        print("Result: ", subtract(num1, num2))
    elif choice == '3':
        print("Result: ", multiply(num1, num2))
    elif choice == '4':
        print("Result: ", divide(num1, num2))
    else:
        print("Помилка. Будь ласка, введіть один із запропонованих варіантів.")

calculator()
```

# Парабола мовою Python

Спочатку ми генеруємо дані для параболи, потім створюємо графік і виводимо його на екран.

```python
import matplotlib.pyplot as plt
import numpy as np


x = np.linspace(-10, 10, 100)
y = x**2


plt.plot(x, y, label='Парабола: $y = x^2$')

plt.title('Графік параболи')
plt.xlabel('x-вісь')
plt.ylabel('y-вісь')
plt.legend()


plt.show()
```
